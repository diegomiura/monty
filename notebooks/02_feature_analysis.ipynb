# Feature analysis
Distributions and sanity checks of the chronological pre-match features.

```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent))
import pandas as pd
feats = pd.read_csv('../data/processed/features.csv', parse_dates=['date'])
feats[['elo_diff','ppm10_a','ew_gf_a','elo_expected_a']].describe()
```

```python
import matplotlib.pyplot as plt
sub = feats[feats['date']>='2000-01-01']
fig, ax = plt.subplots(1,2, figsize=(11,3.5))
ax[0].hist(sub['elo_diff'], bins=60); ax[0].set_title('elo_diff')
sub.groupby(sub['elo_expected_a'].round(1))['outcome'].apply(lambda s:(s==0).mean()).plot(ax=ax[1], marker='o')
ax[1].plot([0,1],[0,1],'k--'); ax[1].set_title('Elo expected vs observed win rate')
plt.show()
```
